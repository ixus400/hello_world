1. 当视频播放或暂停时，手指在屏幕左边上下滑动，改变屏幕亮度；在右边上下滑动，改变音量；底部左右滑动改变播放位置；顶部左右滑动，改变播放速度（按步长改变，默认步长是0.25， 播放速度最小为0.25， 最大为4， 步长可在后面提到的'设置'页面里由用户调整）。
2. 当视频播放或暂停时，轻触屏幕，则显示控制面板，包括：暂停，停止，快进（10秒），快退（10秒），进度条（灰色显示），这些出现在屏幕的底部；在屏幕的左边，出现四个按钮：'A'按钮，按下后，当前的播放位置设为AB段播放的起始点；'B' 按钮，按下后，当前的播放位置+10秒设置为AB段播放的终止点，如果种植点超过视频的终点，则将视频的终点设为'B'点。进度条上AB点位置出现相应的两个拖拽按钮，手指拖拽任意按钮沿进度条方向移动，可调整AB点位置。进度条AB段颜色变为深棕色，如果当前播放位置在A点之后，则A点到当前播放点进度条部分颜色为亮橙色。另外两个按钮为速度'+'，和'-'键：点按'+'则播放速度增加一个步长，默认为0.25（即：当前播放速度为1时，点按'+'后，播放速度变为1.25，再点按后，播放速度增加为1.5，最大为4，到达4后，再点按'+'，速度不再增加，此时无需提示）；相反，点按'-'，则播放速度减少一个步长，默认0.25(即：当前播放速度为1时，点按'-'后，播放速度减为0.75，再点按后，播放速度减为0.5，最小播放速度为0.25，当播放速度到达最小值后，再点按'-'，速度不再减小，此时无需提示）。'A','B','+','-'四个按钮大小为16x16个像素点，其中'A'和'B'按钮背景为深棕色，字母A,B为橙色，背景与字母均半透明（透明度75%）；'+','-'按钮背景为透明灰色，字符'+'和'-'为白色，同样也是透明（透明度75%）。
3. 当视频播放或暂停时，屏幕左边中间出现一个锁屏按钮，按下后屏幕及手机的物理按键都不能改变屏幕的状态和播放器的状态。
4. 当视频播放或暂停时，轻触屏幕和手机按键，如音量键，拍照键，电源键等则在屏幕左上方显示：当前播放速度，AB点的位置信息。调整速度时（手指在屏幕上方左右滑动，或点按'+'和'-'按钮，显示实时更新，同样，拖拽AB滑块，或点按'A''B'按钮，显示也实时更新。播放速度显示字体颜色为白色，AB点位置信息显示为橙色。
5. 当视频播放或暂停时，在屏幕左上方的显示信息下面，设一个定时按钮(按钮为闹钟型),按下这个定时按钮，则出现一个半透明的滑块，用来调整定时退出时间，时间滑块刻度设置为15， 30， 45， 60，90，120等刻度值，拖动滑块，定时时长只在这些固定值变动(单位为分钟)。滑块设置好后2秒钟，自动消失，定时时长则固定显示在定时按钮的右边。
6. 从视频顶部向下滑屏，出现设置页面，包括 a. 设置播放速度步长：0.1， 0.2， 0.25 三种; b. 设置息屏播放: yes, no， 当用户选'yes'后，按下手机息屏键，播放器继续只播放音频， 用户若选'no'，则息屏后播放器暂停播放，处于息屏状态，用户恢复屏显后，可按控制面板的播放按钮继续播放；c. 横屏竖屏自动调整: yes, no, 用户选'yes'，则屏幕自动适应横屏和竖屏，用户选'no'，则播放器屏幕保持正常播放状态，不随屏幕旋转而改变方向（宽视频则始终横屏播放，窄视频始终竖屏播放）；d. 视频屏幕调整：1. 按比例最大显示（即保持视频长宽比不变，按屏幕允许的最大屏幕播放）2. 拉伸适应屏幕，视频按手机屏幕允许的最大屏幕拉伸播放，不再保持原比例）。
